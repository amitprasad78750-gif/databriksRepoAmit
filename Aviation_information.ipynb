```python
from pyspark.sql.types import StructType, StructField
from pyspark.sql.types import IntegerType, StringType, DateType

schema = StructType([
    StructField("YEAR", IntegerType(), True),
    StructField("MONTH_NUM", IntegerType(), True),
    StructField("MONTH_MON", StringType(), True),
    StructField("FLT_DATE", DateType(), True),
    StructField("APT_ICAO", StringType(), True),
    StructField("APT_NAME", StringType(), True),
    StructField("STATE_NAME", StringType(), True),
    StructField("FLT_DEP_1", IntegerType(), True),
    StructField("FLT_ARR_1", IntegerType(), True),
    StructField("FLT_TOT_1", IntegerType(), True),
    StructField("FLT_DEP_IFR_2", IntegerType(), True),
    StructField("FLT_ARR_IFR_2", IntegerType(), True),
    StructField("FLT_TOT_IFR_2", IntegerType(), True)
])
```

```python
df = spark.read \
    .option("header", "true") \
    .schema(schema) \
    .csv("/Volumes/youtube/default/mytube/kaggle/airport_traffic_2025.csv")
```
